# 15.Pie chart

```python
plt.figure(figsize=(10,10))
plt.pie(data['type'].value_counts(),
        labels=data['type'].value_counts().index,autopct='%1.2f%%');
```

# 16.Filtering

```python
data[(data['type']=='Movie') & (data['release_year']==2020)] # To show movies released in the year 2020.
```

```python
mtii=data[((data['type']=='Movie')|(data['type']=='TV Show')) & (data['country']=='India')] # To show TV show and movies released in the India.
mtii
```

```python
plt.figure(figsize=(10,10)) # Pie Chart to show how many percent Indian watched Movie and TV Show .
plt.pie(mtii['type'].value_counts(),
        labels=mtii['type'].value_counts().index,autopct='%1.2f%%');
```

# 20.Number of title  vs  release_year

```python
plt.figure(figsize=(60,40))
sns.countplot(x=data['release_year'],order=data['release_year'].value_counts().index.sort_values())
plt.xlabel('release_year')
plt.ylabel('No: of release_year')
plt.title('Number of title  vs  release_year')
```
